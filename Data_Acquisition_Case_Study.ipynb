## Q6. Write code to get the list of files in your Downloads folder and save it to a CSV file name "download_list.csv"

```python
import os
import pandas as pd

# Get the path to Downloads folder
downloads_path = os.path.join(os.path.expanduser("~"), "Downloads")

# Get a list of files (not folders) in the Downloads directory
files = [f for f in os.listdir(downloads_path) if os.path.isfile(os.path.join(downloads_path, f))]

# Create a DataFrame with the file names
df = pd.DataFrame(files, columns=["File Name"])

# Define the output CSV file path
output_csv_path = os.path.join(downloads_path, "download_list.csv")

# Save the DataFrame to CSV
df.to_csv(output_csv_path, index=False)

print(f"List of files saved to: {output_csv_path}")
```

## Q13. Identify any potential outliers in a numerical column of your choice. Explain your approach.


```python
col = 'windspeed'

# Calculate Q1, Q3, and IQR
Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1

# Calculate bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

print(f"Number of outliers in '{col}':", len(outliers))
print("Outlier windspeed values:")
print(outliers[[col]].drop_duplicates().sort_values(by=col))
```

## Q.16 In which season (Spring, Summer, Fall, Winter) people rented bikes the most?

```python
season_mapping = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
df['season_name'] = df['season'].map(season_mapping)

# Group by season and sum the total rentals
season_rentals = df.groupby('season_name')['cnt'].sum().sort_values(ascending=False)

print("Total rentals per season:",season_rentals)

# Find the season with the highest rentals
most_rented_season = season_rentals.idxmax()
print(f" Season with the highest bike rentals: {most_rented_season}")
```

## Q.17 What is the peak hour in which bike rents the most?

```python
import pandas as pd

# Load the hourly dataset
df = pd.read_csv(r"C:\Users\rahul\OneDrive\Desktop\anitha\DSA\dsa assignments\data acquisition case study\bike+sharing+dataset\hour.csv")

# Group by 'hr' (hour of the day) and sum total rentals
hourly_rentals = df.groupby('hr')['cnt'].sum()

# Find the hour with maximum total rentals
peak_hour = hourly_rentals.idxmax()
peak_count = hourly_rentals.max()

print(" Peak Hour:", peak_hour)
print("Total Rentals at Peak Hour:", peak_count)
```

## Q.18 In which day of a week bikes rents out most?

```python
import pandas as pd

# Load the dataset
df = pd.read_csv(r"C:\Users\rahul\OneDrive\Desktop\anitha\DSA\dsa assignments\data acquisition case study\bike+sharing+dataset\day.csv")

# Map weekday numbers to names
weekday_map = {
    0: 'Sunday', 1: 'Monday', 2: 'Tuesday', 3: 'Wednesday',
    4: 'Thursday', 5: 'Friday', 6: 'Saturday'
}
df['weekday_name'] = df['weekday'].map(weekday_map)

# Group by day of the week and sum total rentals
weekday_rentals = df.groupby('weekday_name')['cnt'].sum().sort_values(ascending=False)

print("Total rentals by weekday:")
print(weekday_rentals)

# Identify the day with the most rentals
peak_day = weekday_rentals.idxmax()
print(f" Day with the highest bike rentals: {peak_day}")
```

## Q.20 What is the maximum temperature observed in each of the seasons?

```python
df = pd.read_csv(r"C:\Users\rahul\OneDrive\Desktop\anitha\DSA\dsa assignments\data acquisition case study\bike+sharing+dataset\day.csv")

season_mapping = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
df['season_name'] = df['season'].map(season_mapping)

# Compute max temperature in each season (normalized)
max_temp_normalized = df.groupby('season_name')['temp'].max()

# Convert to actual temperature by multiplying by 41 (as per dataset documentation)
max_temp_celsius = max_temp_normalized * 41

# Display results
print(" Maximum Observed Temperature (C) by Season:")
print(max_temp_celsius.round(2))
```
